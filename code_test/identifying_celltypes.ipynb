```python
import pyBigWig
for j,identifier in enumerate(targets):
    bw_file = f'/data1/lesliec/sarthak/data/DK_zarr/bw_files/{identifier}.bigWig'
    values = pyBigWig.open(bw_file).values(chrom, start, end)
    values = np.array(values, dtype=np.float16)
    values = np.nan_to_num(values) #by doing nan to num after, anything that is infinity is set to a large number
    
    # print(j)
    data[j] = values
```
