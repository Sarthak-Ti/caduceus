```python
import pandas as pd
import pyBigWig
final_df = pd.read_csv('/data1/lesliec/sarthak/data/DK_zarr/targets_DNase_immune.txt', sep='\t')
bw_files = final_df['file'].to_list()
bws = {}
for i,bw in enumerate(bw_files):
    bws[i] = pyBigWig.open(bw)
#/data1/lesliec/sarthak/data/DK_zarr/bw_files/ENCFF207ATZ.bigWig
```

```python
#and the zarr
import zarr
zo = zarr.open('/data1/lesliec/sarthak/data/DK_zarr/zarr_arrays/dnase_immune.zarr', mode='r')
zo.tree()
```

```python
#let's also open the blacklist and unmappable regions
import numpy as np
unmappable_file: str = '/data1/lesliec/sarthak/data/borzoi/umap.npz'
blacklist_file: str = '/data1/lesliec/sarthak/data/borzoi/blacklist.npz'

with np.load(blacklist_file) as data:
    blacklist = {key: np.array(data[key]) for key in data}
with np.load(unmappable_file) as data:
    unmappable = {key: np.array(data[key]) for key in data}
```

```python
#let's do a loop
from tqdm import tqdm
celltype = 0
for chrom in unmappable.keys():
    for start in tqdm(range(0,len(unmappable[chrom]),1_000_000)):
        stop = start+1_000_000
        if stop > len(unmappable[chrom]):
            stop = len(unmappable[chrom])
        for celltype in range(24):
            bw_values = np.array(bws[celltype].values(chrom, start, stop))
            extreme_clip = np.percentile(bw_values, 100*0.9999999)
            diff = bw_values - zo[chrom][celltype,start:stop]/2
            # indexes = np.where(diff != 0)[0]
            
            #now we remove values of diff that are above extreme clip
            diff[bw_values > extreme_clip] = 0
            
            #and then we remove values that are unmappable
            diff[unmappable[chrom][start:stop] == 1] = 0
            diff[blacklist[chrom][start:stop] == 1] = 0
            
            if (diff > 0.1).any():
                print(f"diff not zero for {chrom} {start} {stop} {celltype}")
                print(diff)
                assert False
```

```python
#oh extreme clip is caclulated after the unmappable and blacklist are removed, let's try again!

for chrom in unmappable.keys():
    for start in tqdm(range(0,len(unmappable[chrom]),1_000_000)):
        stop = start+1_000_000
        if stop > len(unmappable[chrom]):
            stop = len(unmappable[chrom])
        for celltype in range(24):
            bw_values = np.array(bws[celltype].values(chrom, start, stop))
            bw_values = np.nan_to_num(bw_values)
            
            blist_region = blacklist[chrom][start:stop]
            umap_region = unmappable[chrom][start:stop]
            
            bw_values[blist_region == 1] = 0
            bw_values[umap_region == 1] = 0
            
            extreme_clip = np.percentile(bw_values, 100*0.9999999)
            extreme_clip = min(extreme_clip, 1000)
            diff = bw_values - zo[chrom][celltype,start:stop]
            # indexes = np.where(diff != 0)[0]
            
            #now we remove values of diff that are above extreme clip
            diff[bw_values > extreme_clip] = 0
            
            diff[blist_region == 1] = 0
            diff[umap_region == 1] = 0
            
            if (diff > 0.5).any():
                print(f"diff not zero for {chrom} {start} {stop} {celltype}")
                print(diff)
                assert False
```
