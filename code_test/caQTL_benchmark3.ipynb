```python
chrom='chr1'
pos=817186-1
start = pos - 524288//2
end = start + 524288
evals.dataset.expand_seqs(chrom,start,end)
```

```python
import torch
mapping = {
        'A': torch.tensor([1, 0, 0, 0], dtype=torch.float32),
        'C': torch.tensor([0, 1, 0, 0], dtype=torch.float32),
        'G': torch.tensor([0, 0, 1, 0], dtype=torch.float32),
        'T': torch.tensor([0, 0, 0, 1], dtype=torch.float32),
        'N': torch.tensor([0, 0, 0, 0], dtype=torch.float32),
    }

onehot_mapping = {0:'A',1:'C',2:'G',3:'T'}
```

```python
#let's see what a prediction looks like
# si = 524288//2 - 250
# se = 524288//2 + 250
import matplotlib.pyplot as plt
idx = 1937
mask_size = 500 #seems even better!!
si = 524288//2 - int(mask_size/2)
se = 524288//2 + int(mask_size/2)
((s,a),(su,au)) = evals.dataset[idx]

data = (None,None,su,au)
out = evals.mask(si,se, data=data, mask_accessibility=True)
current_nuc = out[2][524288//2].cpu().numpy()
current_nuc = np.argmax(current_nuc)
current_nuc = onehot_mapping[current_nuc]

if current_nuc != 'G':
    raise ValueError(f'Expected G, got {current_nuc}')
alt_key = 'A'

data[2][524288//2,:4] = mapping[alt_key]

out2 = evals.mask(si,se, data=data, mask_accessibility=True)

pred1 = out[1][0, 524288//2-350:524288//2+350, 0].cpu().numpy()
pred2 = out2[1][0, 524288//2-350:524288//2+350, 0].cpu().numpy()

plt.plot(pred1, label='Model Predictions', color='orange')
plt.plot(pred2, label='Model Predictions Modified', color='blue')
plt.plot(au[524288//2-350:524288//2+350,0], label='True data', color='red')
plt.plot(out[-1][0, 0, 524288//2-350:524288//2+350].cpu().numpy(), label='Model input', color='green')

plt.legend()
plt.xlabel('position')
plt.ylabel('value')
plt.title(f'model changes mask size {mask_size}')
```

```python
#first let's see about running the model, does bigger batch even help?
import os
os.environ['CUDA_VISIBLE_DEVICES'] = '2'
import sys
sys.path.append('/data1/lesliec/sarthak/caduceus/')
from evals.evals_utils_joint import Evals
ckpt_path = '/data1/lesliec/sarthak/caduceus/outputs/2025-04-28/15-26-30-700432/checkpoints/14-val_loss=0.27260.ckpt'
evals = Evals(ckpt_path, data_idxs=None, data_path='/data1/lesliec/sarthak/data/DK_zarr/zarr_arrays/cell_type_arrays/GM12878_DNase.npz')
batch0 = evals.dataset[0]
x = batch0[0][0].unsqueeze(0)
y = batch0[0][1].unsqueeze(0)
print(x.shape, y.shape)
```

```python
x,y = x.to('cuda'), y.to('cuda')
```

```python
from tqdm import tqdm
import torch
torch.cuda.synchronize()
for i in tqdm(range(160)):
    with torch.no_grad():
        x1 = evals.encoder(x,y)
        x1,_ = evals.backbone(x1)
        x1 = evals.decoder(x1)
        seq,acc = x1
#now sync cuda
torch.cuda.synchronize()
#4 minutes 7 seconds
```
