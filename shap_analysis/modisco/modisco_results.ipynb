```python
import re
import os

# Define the directory containing your HTML files
directory = '/path/to/your/html/files'  # Adjust this to the path of your HTML files

# Regex to find and modify the paths
pattern = re.compile(r'(_report(\d+))trimmed')
print(pattern)
testpattern = 'thisisatest_report600trimmed'

# Function to add a slash after the report number
def add_slash(match):
    return f"{match.group(1)}/trimmed"

out = re.sub(pattern,add_slash,testpattern)
print(out)

directory = '/data/leslie/sarthak/hyena/hyena-dna/shap_analysis/modisco/all_celltypes/reports/'
# Loop through each file in the directory
for subdir in os.listdir(directory):
    print(subdir)
```

# now let's create the heatmaps!

```python
#we need to load in the files of all the different celltypes, the idea is we load in the html,  make a list and then threshold them all
#then we make that a value of one array, this approach is the best when it comes to making the values of  a heatmap
directory = '/data/leslie/sarthak/hyena/hyena-dna/shap_analysis/modisco/all_celltypes/reports/'
for subdir in os.listdir(directory):
    tables = pd.read_html(directory+subdir+'/'+'motifs.html')
    df = tables[0]
    print(df.shape)
    #now make the 3 lists that are paired
    qvals = pd.concat([df['qval0'],df['qval1'],df['qval2']], axis=0).to_numpy()
    matches = pd.concat([df['match0'],df['match1'],df['match2']], axis=0).to_numpy()
    seqs = pd.concat([df['num_seqlets'],df['num_seqlets'],df['num_seqlets']], axis=0).to_numpy()
    break
```

```python
#then we can threshold the q vals to then find them all
newq = qvals[qvals<.05]
print(newq.shape, newq)
newmatch = matches[qvals<.05]
print(newmatch.shape, newmatch)
newseqs = seqs[qvals<.05]
print(newseqs.shape, newseqs)
```

```python
#from here we can easily make a dictionary that takes the matches to turn it into a new index
match_dict = {}
```

```python
#we need to load in the files of all the different celltypes, the idea is we load in the html,  make a list and then threshold them all
#then we make that a value of one array, this approach is the best when it comes to making the values of  a heatmap
directory = '/data/leslie/sarthak/hyena/hyena-dna/shap_analysis/modisco/all_celltypes/reports/'
dict_value = 0
match_dict = {}
#we don't know how long ot make the array yet, so for now we can just make it a list
seqlets_list = []
qvals_list = []
match_list = []
for subdir in os.listdir(directory):
    if not 'ctst' in subdir: #we only want the ctst ones, this ignores multitasking ones
        continue
    tables = pd.read_html(directory+subdir+'/'+'motifs.html')
    df = tables[0]
    # print(df.shape)
    #now make the 3 lists that are paired
    qvals = pd.concat([df['qval0'],df['qval1'],df['qval2']], axis=0).to_numpy()
    matches = pd.concat([df['match0'],df['match1'],df['match2']], axis=0).to_numpy()
    seqs = pd.concat([df['num_seqlets'],df['num_seqlets'],df['num_seqlets']], axis=0).to_numpy()
    newq = qvals[qvals<.05]
    newmatch = matches[qvals<.05]
    newseqs = seqs[qvals<.05]
    #now we can make a dictionary that takes the matches to turn it into a new index
    for match in newmatch:
        if match not in match_dict:
            match_dict[match] = dict_value
            dict_value += 1
    match_list.append(newmatch)
    qvals_list.append(newq)
    seqlets_list.append(newseqs)
```

```python
print(len(match_list), len(qvals_list), len(seqlets_list))
#loks good!
```

```python
print(match_list[0])
print(qvals_list[0])
```

```python
#now we need to make the array, we can make it a numpy array
import numpy as np
full_array = np.zeros((len(match_dict), len(qvals_list))) #this is num_unique motifs to num_celltypes
print(full_array.shape)
```

```python
#and now we populate it with the values, and the value is dependent on the celltype and the motif
#definitely some kind of for loop, let's go ahead and go through cells
for cell_idx in range(len(qvals_list)):
    #we are iterating through the celltypes
    for counter,match in enumerate(match_list[cell_idx]):
        #this iterates through the matches, then we find the index by the match_dict
        match_idx = match_dict[match]
        # print(match_idx)
        # print(cell_idx)
        #then we can set the value
        full_array[match_idx, cell_idx] = qvals_list[cell_idx][counter]
```
